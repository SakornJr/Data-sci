```python
import matplotlib.pyplot as plt
```

```python
players = ['Ronaldo', 'Messi', 'Neymar', 'Mbappe', 'Salah']
goals = [983, 750, 400, 300, 250]

plt.bar(players, goals, color='blue')
plt.title('Top Football Players by Goals')
plt.xlabel('Players')
plt.ylabel('Goals Scored')

plt.show()
```
